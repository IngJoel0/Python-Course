## Segmento 3: Mostrar mensajes con `print()`

`print()` sirve para mostrar información en pantalla.

Lo que queremos mostrar se escribe dentro de los paréntesis.
Si es texto, lo ponemos entre comillas.

```python
print("Estoy aprendiendo Python")
print("Esta es mi primera clase")
print("Python puede ser divertido")
```

## Segmento 4: Comentarios

Los comentarios son notas para nosotros.  
Python los ignora al ejecutar el programa.

Un comentario empieza con `#`.

Sirven para explicar qué hace una parte del código.

```python
# Este programa muestra un saludo
print("Hola, estudiante")

# Esta línea muestra una frase motivadora
print("Cada error ayuda a aprender")
```

## Segmento 5: Variables

Una variable es como una cajita con nombre donde guardamos información.

Por ejemplo, podemos guardar un nombre, una edad o una ciudad.

<div align="center">
  
</div>

```python
nombre = "Ana"
edad = 14
ciudad = "Tijuana"

print(nombre)
print(edad)
print(ciudad)
```

Podemos usar variables junto con texto.

```python
nombre = "Luis"
print("Hola", nombre)
```

## Segmento 6: Tipos de datos básicos

Python puede guardar diferentes tipos de datos.

Los más importantes al inicio son:

- `str`: texto.
- `int`: número entero.
- `float`: número decimal.
- `bool`: verdadero o falso.

No te preocupes si los nombres parecen raros. Los iremos usando poco a poco.

```python
texto = "Python"
numero_entero = 15
numero_decimal = 9.5
valor_logico = True

print(type(texto))
print(type(numero_entero))
print(type(numero_decimal))
print(type(valor_logico))
```

## Segmento 7: Pedir datos con `input()`

`input()` permite que el usuario escriba información.

Esto hace que el programa sea más interactivo.

```python
nombre = input("¿Cómo te llamas? ")
print("Hola", nombre)
```

Podemos pedir varios datos y después mostrarlos juntos.

```python
nombre = input("Nombre: ")
edad = input("Edad: ")
hobby = input("Hobby favorito: ")

print("----- PERFIL -----")
print("Nombre:", nombre)
print("Edad:", edad)
print("Hobby:", hobby)
```

## Segmento 8: Actividad guiada

Vamos a crear una tarjeta de presentación digital.

La computadora preguntará datos y después los mostrará ordenados.

```python
nombre = input("Escribe tu nombre: ")
edad = input("Escribe tu edad: ")
ciudad = input("Escribe tu ciudad: ")
comida = input("Escribe tu comida favorita: ")

print("==============================")
print("MI TARJETA DE PRESENTACIÓN")
print("==============================")
print("Nombre:", nombre)
print("Edad:", edad)
print("Ciudad:", ciudad)
print("Comida favorita:", comida)
print("==============================")
```
