# Clase 5: Funciones y organización del código

En esta clase aprenderemos a usar funciones.

Las funciones ayudan a organizar el código y evitar repetir instrucciones.

<div align="center">
  
</div>

## Segmento 1: ¿Por qué usar funciones?

Cuando un programa crece, repetir código se vuelve incómodo.

Las funciones nos permiten guardar un bloque de instrucciones con un nombre.
Luego podemos usar ese bloque muchas veces.

````python
print("Hola, Ana")
print("Bienvenida al curso")

print("Hola, Luis")
print("Bienvenido al curso")

print("Hola, Sofía")
print("Bienvenida al curso")
````

Ese código funciona, pero se repite mucho.  
Con una función podemos hacerlo mejor.

## Segmento 2: Crear una función con `def`

Para crear una función usamos `def`.

La estructura básica es:

```python
def nombre_de_la_funcion():
    instrucciones
```

````python
def saludar():
    print("Hola")
    print("Bienvenido al curso de Python")

saludar()
````

## Segmento 3: Llamar una función varias veces

Una función no se ejecuta sola.  
Para usarla, escribimos su nombre con paréntesis.

````python
def saludar():
    print("Hola, programador")
    print("Sigue practicando")
    print("---------------")

saludar()
saludar()
saludar()
````

## Segmento 4: Parámetros

Un parámetro es un dato que la función necesita para trabajar.

Ejemplo: una función para saludar necesita saber el nombre.

<div align="center">
  
</div>

````python
def saludar_persona(nombre):
    print("Hola", nombre)

saludar_persona("Ana")
saludar_persona("Luis")
saludar_persona("Sofía")
````

## Segmento 5: Varios parámetros

Una función puede recibir varios datos.

````python
def mostrar_perfil(nombre, edad, hobby):
    print("----- PERFIL -----")
    print("Nombre:", nombre)
    print("Edad:", edad)
    print("Hobby:", hobby)

mostrar_perfil("Carlos", 15, "fútbol")
````

## Segmento 6: Funciones con `input()`

Podemos pedir datos al usuario y mandarlos a una función.

````python
def tarjeta(nombre, edad, ciudad):
    print("===================")
    print("TARJETA PERSONAL")
    print("===================")
    print("Nombre:", nombre)
    print("Edad:", edad)
    print("Ciudad:", ciudad)

nombre_usuario = input("Nombre: ")
edad_usuario = input("Edad: ")
ciudad_usuario = input("Ciudad: ")

tarjeta(nombre_usuario, edad_usuario, ciudad_usuario)
````

## Segmento 7: Funciones que calculan

Las funciones también pueden hacer operaciones.

````python
def sumar(a, b):
    resultado = a + b
    print("Resultado:", resultado)

sumar(5, 3)
sumar(10, 20)
````

## Segmento 8: `return`

`return` devuelve un resultado.

Esto permite guardar el resultado en una variable y usarlo después.

````python
def sumar(a, b):
    return a + b

resultado = sumar(5, 3)

print("El resultado guardado es:", resultado)
````

## Segmento 9: Diferencia entre `print()` y `return`

`print()` solo muestra algo.

`return` devuelve un valor para seguir trabajando con él.

````python
def sumar_con_print(a, b):
    print(a + b)

resultado = sumar_con_print(5, 3)
print("Resultado guardado:", resultado)
````

````python
def sumar_con_return(a, b):
    return a + b

resultado = sumar_con_return(5, 3)
print("Resultado guardado:", resultado)
````

## Segmento 10: Operaciones básicas con funciones

Vamos a crear funciones para una calculadora.

````python
def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    if b == 0:
        return "No se puede dividir entre cero"
    else:
        return a / b

print(sumar(10, 5))
print(restar(10, 5))
print(multiplicar(10, 5))
print(dividir(10, 5))
````

## Segmento 11: Funciones con texto

También podemos crear funciones que trabajen con palabras.

````python
def convertir_mayusculas(texto):
    return texto.upper()

mensaje = input("Escribe un mensaje: ")

print(convertir_mayusculas(mensaje))
````

````python
def contar_letras(palabra):
    return len(palabra)

palabra = input("Escribe una palabra: ")

print("Cantidad de letras:", contar_letras(palabra))
````

## Segmento 12: Funciones con listas

Las funciones también pueden recibir listas.

````python
def mostrar_lista(lista):
    for elemento in lista:
        print("-", elemento)

hobbies = ["dibujar", "jugar", "leer", "programar"]

mostrar_lista(hobbies)
````

````python
def calcular_promedio(calificaciones):
    return sum(calificaciones) / len(calificaciones)

notas = [8, 9, 10, 7, 9]

print("Promedio:", calcular_promedio(notas))
````

## Segmento 13: Menú con funciones

Podemos separar el menú en funciones para que el programa se vea más limpio.

<div align="center">
  
</div>

````python
def mostrar_menu():
    print("----- MENÚ -----")
    print("1. Saludar")
    print("2. Mostrar frase")
    print("3. Salir")

def saludar():
    nombre = input("Nombre: ")
    print("Hola", nombre)

def mostrar_frase():
    print("La práctica hace al programador")

opcion = ""

while opcion != "3":
    mostrar_menu()
    opcion = input("Elige una opción: ")

    if opcion == "1":
        saludar()
    elif opcion == "2":
        mostrar_frase()
    elif opcion == "3":
        print("Programa terminado")
    else:
        print("Opción no válida")
````

## Segmento 14: Funciones en juegos

Las funciones permiten dividir un juego en partes pequeñas.

````python
def revisar_respuesta(respuesta, correcta):
    if respuesta.lower() == correcta.lower():
        return 1
    else:
        return 0

puntos = 0

respuesta = input("¿Qué lenguaje estamos aprendiendo? ")
puntos += revisar_respuesta(respuesta, "Python")

respuesta = input("¿Qué función muestra texto en pantalla? ")
puntos += revisar_respuesta(respuesta, "print")

print("Puntos:", puntos)
````

## Segmento 15: Generador de personaje

Este ejemplo mezcla listas, `random` y funciones.

````python
import random

def elegir_nombre():
    nombres = ["Luna", "Max", "Kai", "Nova"]
    return random.choice(nombres)

def elegir_poder():
    poderes = ["volar", "ser invisible", "controlar el fuego", "teletransportarse"]
    return random.choice(poderes)

def crear_personaje():
    nombre = elegir_nombre()
    poder = elegir_poder()

    print("Personaje:", nombre)
    print("Poder:", poder)

crear_personaje()
````
