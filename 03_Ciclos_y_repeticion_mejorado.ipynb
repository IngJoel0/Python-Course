# Clase 3: Ciclos y repetición

En esta clase aprenderemos a repetir instrucciones.

Los ciclos sirven para no escribir lo mismo muchas veces.

## Segmento 1: ¿Por qué usar ciclos?

Sin ciclos, tendríamos que repetir líneas manualmente.

```python
print("Estoy aprendiendo Python")
print("Estoy aprendiendo Python")
print("Estoy aprendiendo Python")
print("Estoy aprendiendo Python")
print("Estoy aprendiendo Python")
```

Con un ciclo, podemos hacerlo en menos líneas.

```python
for i in range(5):
    print("Estoy aprendiendo Python")
```

## Segmento 2: Ciclo `for`

Usamos `for` cuando sabemos cuántas veces queremos repetir algo.

`range(5)` repite 5 veces, contando desde 0 hasta 4.

```python
for i in range(5):
    print(i)
```

Podemos indicar inicio y final.

```python
for numero in range(1, 6):
    print(numero)
```

También podemos avanzar de 2 en 2.

```python
for numero in range(2, 11, 2):
    print(numero)
```

## Segmento 3: Tabla de multiplicar

Este ejemplo muestra por qué los ciclos son útiles.

```python
numero = int(input("¿Qué tabla quieres ver? "))

for i in range(1, 11):
    print(numero, "x", i, "=", numero * i)
```

## Segmento 4: Contadores

Un contador es una variable que va aumentando.

Se usa para contar intentos, puntos o repeticiones.

<div align="center">
  
</div>

```python
contador = 0

for i in range(5):
    contador += 1
    print("Contador:", contador)
```

## Segmento 5: Acumuladores

Un acumulador va sumando valores.

Sirve para totales, puntos o promedios.

```python
total = 0

for i in range(3):
    numero = int(input("Escribe un número: "))
    total += numero

print("Total:", total)
```

## Segmento 6: Promedio de calificaciones

Pediremos varias calificaciones y calcularemos el promedio.

```python
total = 0

for i in range(3):
    calificacion = float(input("Escribe una calificación: "))
    total += calificacion

promedio = total / 3
print("Promedio:", promedio)
```

## Segmento 7: Ciclo `while`

Usamos `while` cuando queremos repetir mientras una condición se cumpla.

Por ejemplo: repetir hasta que el usuario escriba `salir`.

```python
contador = 1

while contador <= 5:
    print("Número:", contador)
    contador += 1
```

## Segmento 8: Repetir hasta escribir salir

Este tipo de programa es común en menús.

```python
respuesta = ""

while respuesta != "salir":
    respuesta = input("Escribe algo o escribe salir: ").lower()
    print("Escribiste:", respuesta)

print("Programa terminado")
```

## Segmento 9: Menú interactivo

Un menú permite que el usuario elija qué quiere hacer.

<div align="center">
  
</div>

```python
opcion = ""

while opcion != "4":
    print("----- MENÚ -----")
    print("1. Saludar")
    print("2. Mostrar frase")
    print("3. Decir la hora de practicar")
    print("4. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        print("Hola")
    elif opcion == "2":
        print("Python se aprende practicando")
    elif opcion == "3":
        print("Es hora de practicar")
    elif opcion == "4":
        print("Adiós")
    else:
        print("Opción no válida")
```

## Segmento 10: `break`

`break` detiene un ciclo antes de que termine.

Sirve cuando ya conseguimos lo que queríamos.

```python
for i in range(10):
    print(i)

    if i == 5:
        print("Llegamos al 5, detenemos el ciclo")
        break
```

## Segmento 11: `continue`

`continue` salta una vuelta del ciclo.

En este ejemplo no imprimimos el número 3.

```python
for i in range(1, 6):
    if i == 3:
        continue

    print(i)
```

## Segmento 12: Juego con intentos

Mejoraremos el juego de adivinar el número usando ciclos.

```python
import random

numero_secreto = random.randint(1, 10)
intentos = 3

for i in range(intentos):
    intento = int(input("Adivina un número del 1 al 10: "))

    if intento == numero_secreto:
        print("¡Ganaste!")
        break
    elif intento < numero_secreto:
        print("Muy bajo")
    else:
        print("Muy alto")

print("El número era:", numero_secreto)
```
