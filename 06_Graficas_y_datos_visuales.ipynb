## Segmento 1: ¿Para qué sirven las gráficas?

Una gráfica sirve para ver datos de manera visual.

Por ejemplo, si tenemos calificaciones, ventas, votos o gustos del grupo, una gráfica puede ayudarnos a ver:

- Qué opción ganó.
- Qué dato es más alto.
- Qué dato es más bajo.
- Cómo cambian los datos.
- Comparaciones rápidas.

En programación, no todo tiene que verse como texto. También podemos mostrar resultados visuales.

## Segmento 2: Importar `matplotlib`

Para crear gráficas usaremos `matplotlib.pyplot`.

Normalmente se importa así:

```python
import matplotlib.pyplot as plt
```

El nombre `plt` es un apodo corto para usar la biblioteca más fácil.

````python
import matplotlib.pyplot as plt

print("Matplotlib está listo")
````

## Segmento 3: Primera gráfica de barras

Una gráfica de barras sirve para comparar cantidades.

Ejemplo: materias y calificaciones.

<div align="center">
  
</div>

````python
import matplotlib.pyplot as plt

materias = ["Mate", "Español", "Historia", "Ciencias"]
calificaciones = [9, 8, 10, 7]

plt.bar(materias, calificaciones)
plt.title("Mis calificaciones")
plt.xlabel("Materias")
plt.ylabel("Calificación")
plt.show()
````

### Explicación del código

- `materias` guarda los nombres que aparecen abajo.
- `calificaciones` guarda los valores de cada barra.
- `plt.bar()` crea la gráfica de barras.
- `plt.title()` agrega el título.
- `plt.xlabel()` nombra el eje horizontal.
- `plt.ylabel()` nombra el eje vertical.
- `plt.show()` muestra la gráfica.

## Segmento 4: Cambiar los datos de una gráfica

Lo interesante es que podemos cambiar los datos y la gráfica cambia automáticamente.

Prueba modificar los nombres y los números.

````python
import matplotlib.pyplot as plt

comidas = ["Pizza", "Tacos", "Sushi", "Hamburguesa"]
votos = [5, 8, 3, 6]

plt.bar(comidas, votos)
plt.title("Comidas favoritas del grupo")
plt.xlabel("Comida")
plt.ylabel("Votos")
plt.show()
````

## Segmento 5: Gráfica de líneas

Una gráfica de líneas sirve para mostrar cambios.

Ejemplo:

- Temperatura durante el día.
- Puntos ganados por nivel.
- Progreso de estudio durante una semana.

<div align="center">
  
</div>

````python
import matplotlib.pyplot as plt

dias = ["Lun", "Mar", "Mié", "Jue", "Vie"]
horas_estudio = [1, 2, 1, 3, 2]

plt.plot(dias, horas_estudio, marker="o")
plt.title("Horas de estudio en la semana")
plt.xlabel("Día")
plt.ylabel("Horas")
plt.show()
````

### Explicación del código

- `plt.plot()` crea una gráfica de líneas.
- `marker="o"` pone puntos en cada dato.
- Los días están en el eje horizontal.
- Las horas están en el eje vertical.

## Segmento 6: Gráfica de pastel

Una gráfica de pastel sirve para representar partes de un total.

Ejemplo: porcentaje de gustos del grupo.

````python
import matplotlib.pyplot as plt

actividades = ["Videojuegos", "Deporte", "Música", "Leer"]
votos = [10, 6, 8, 3]

plt.pie(votos, labels=actividades, autopct="%1.1f%%")
plt.title("Actividades favoritas")
plt.show()
````

### ¿Qué significa `autopct`?

`autopct="%1.1f%%"` muestra porcentajes dentro de la gráfica.

No necesitas memorizarlo, solo saber que sirve para mostrar el porcentaje de cada parte.

## Segmento 7: Crear datos con listas

Antes de usar `numpy`, podemos crear datos con listas normales de Python.

````python
numeros = [1, 2, 3, 4, 5]
dobles = []

for numero in numeros:
    dobles.append(numero * 2)

print("Números:", numeros)
print("Dobles:", dobles)
````

````python
import matplotlib.pyplot as plt

numeros = [1, 2, 3, 4, 5]
dobles = [2, 4, 6, 8, 10]

plt.plot(numeros, dobles, marker="o")
plt.title("Tabla del 2")
plt.xlabel("Número")
plt.ylabel("Doble")
plt.show()
````

## Segmento 8: Introducción a `numpy`

`numpy` es una biblioteca muy usada para trabajar con números.

Nos permite crear arreglos numéricos y hacer operaciones más rápido.

Normalmente se importa así:

```python
import numpy as np
```

`np` es un apodo corto para usar `numpy`.

````python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = x * 2

print("x:", x)
print("y:", y)
````

Observa que con `numpy` pudimos multiplicar todos los valores de `x` por 2 en una sola línea.

Con listas normales necesitaríamos un ciclo.

## Segmento 9: Gráfica usando `numpy`

Ahora vamos a usar `numpy` para crear datos y luego graficarlos con `matplotlib`.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = x * 3

plt.plot(x, y, marker="o")
plt.title("Tabla del 3")
plt.xlabel("x")
plt.ylabel("x por 3")
plt.show()
````

## Segmento 10: Crear una secuencia con `np.arange()`

`np.arange()` permite crear una secuencia de números.

Ejemplo:

```python
np.arange(1, 11)
```

Esto crea números del 1 al 10.

````python
import numpy as np

numeros = np.arange(1, 11)

print(numeros)
````

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = x ** 2

plt.plot(x, y, marker="o")
plt.title("Números al cuadrado")
plt.xlabel("Número")
plt.ylabel("Cuadrado")
plt.show()
````

## Segmento 11: Comparar dos líneas

Podemos poner más de una línea en la misma gráfica para comparar datos.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
doble = x * 2
triple = x * 3

plt.plot(x, doble, marker="o", label="Doble")
plt.plot(x, triple, marker="o", label="Triple")

plt.title("Doble y triple de un número")
plt.xlabel("Número")
plt.ylabel("Resultado")
plt.legend()
plt.show()
````

### ¿Qué hace `label` y `legend()`?

- `label` le pone nombre a cada línea.
- `plt.legend()` muestra la leyenda de la gráfica.

Esto ayuda a saber qué representa cada línea.
