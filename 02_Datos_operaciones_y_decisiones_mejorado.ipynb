# Clase 2: Datos, operaciones y decisiones

En esta clase aprenderemos a convertir datos, hacer operaciones y tomar decisiones.

Esto es importante porque muchos programas necesitan responder diferente según lo que escriba el usuario.

## Segmento 1: Repaso de `input()`

`input()` siempre recibe información como texto.

Aunque el usuario escriba `15`, Python lo guarda como `"15"` si usamos `input()` directamente.

```python
edad = input("¿Cuántos años tienes? ")

print("Tu edad es:", edad)
print("Tipo de dato:", type(edad))
```

## Segmento 2: Convertir datos

Para hacer operaciones con números escritos por el usuario, necesitamos convertirlos.

Usamos:

- `int()` para números enteros.
- `float()` para números decimales.

<div align="center">
  
</div>

```python
edad = int(input("¿Cuántos años tienes? "))

print("El próximo año tendrás:", edad + 1)
```

```python
precio = float(input("Precio del producto: "))

print("Precio con 10 pesos extra:", precio + 10)
```

## Segmento 3: Operadores aritméticos

Python puede hacer operaciones matemáticas.

Los operadores más usados son:

- `+` suma.
- `-` resta.
- `*` multiplicación.
- `/` división.
- `//` división entera.
- `%` residuo.
- `**` potencia.

```python
a = 10
b = 3

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
print("División:", a / b)
print("División entera:", a // b)
print("Residuo:", a % b)
print("Potencia:", a ** b)
```

## Segmento 4: Mini calculadora

Ahora combinamos `input()`, conversión y operaciones.

```python
numero1 = float(input("Primer número: "))
numero2 = float(input("Segundo número: "))

print("Suma:", numero1 + numero2)
print("Resta:", numero1 - numero2)
print("Multiplicación:", numero1 * numero2)
print("División:", numero1 / numero2)
```

## Segmento 5: Comparaciones

Las comparaciones sirven para revisar si algo es verdadero o falso.

Por ejemplo:

- ¿La edad es mayor o igual a 14?
- ¿La contraseña es correcta?
- ¿La calificación es mayor a 6?

```python
edad = 15

print(edad >= 14)
print(edad < 14)
print(edad == 15)
print(edad != 18)
```

## Segmento 6: Decisiones con `if`

`if` significa “si”.

Si una condición se cumple, el programa ejecuta cierto código.

<div align="center">
  
</div>

```python
edad = int(input("¿Cuántos años tienes? "))

if edad >= 14:
    print("Puedes tomar el curso")
```

## Segmento 7: `if` y `else`

`else` significa “si no”.

Lo usamos para decir qué pasa cuando la condición no se cumple.

```python
edad = int(input("¿Cuántos años tienes? "))

if edad >= 14:
    print("Puedes tomar el curso")
else:
    print("Todavía no puedes tomar el curso")
```

## Segmento 8: `elif`

`elif` sirve cuando tenemos varias posibilidades.

Ejemplo: calificación excelente, buena, aprobada o reprobada.

```python
calificacion = float(input("Escribe tu calificación: "))

if calificacion == 10:
    print("Excelente")
elif calificacion >= 8:
    print("Muy bien")
elif calificacion >= 6:
    print("Aprobado")
else:
    print("Reprobado")
```

## Segmento 9: Evitar problemas con mayúsculas

Si el usuario escribe `Verde`, `VERDE` o `verde`, puede causar problemas.

Con `.lower()` convertimos todo a minúsculas.

```python
color = input("Color del semáforo: ")
color = color.lower()

if color == "verde":
    print("Avanza")
elif color == "amarillo":
    print("Precaución")
elif color == "rojo":
    print("Detente")
else:
    print("Color no válido")
```

## Segmento 10: Biblioteca `random`

`random` sirve para generar resultados aleatorios.

Es útil para juegos, sorteos y dinámicas.

<div align="center">
  
</div>

```python
import random

numero = random.randint(1, 10)
print("Número aleatorio:", numero)
```

```python
import random

opciones = ["pizza", "tacos", "sushi", "hamburguesa"]
comida = random.choice(opciones)

print("Hoy toca comer:", comida)
```

## Segmento 11: Juego de adivinar número

La computadora elige un número y el jugador intenta adivinarlo.

```python
import random

numero_secreto = random.randint(1, 10)
intento = int(input("Adivina un número del 1 al 10: "))

if intento == numero_secreto:
    print("¡Ganaste!")
elif intento < numero_secreto:
    print("Muy bajo. El número era:", numero_secreto)
else:
    print("Muy alto. El número era:", numero_secreto)
```

## Segmento 12: Piedra, papel o tijera

Este juego usa decisiones y `random`.

```python
import random

opciones = ["piedra", "papel", "tijera"]

jugador = input("Elige piedra, papel o tijera: ").lower()
computadora = random.choice(opciones)

print("Tú elegiste:", jugador)
print("La computadora eligió:", computadora)

if jugador == computadora:
    print("Empate")
elif jugador == "piedra" and computadora == "tijera":
    print("Ganaste")
elif jugador == "papel" and computadora == "piedra":
    print("Ganaste")
elif jugador == "tijera" and computadora == "papel":
    print("Ganaste")
else:
    print("Perdiste")
```
