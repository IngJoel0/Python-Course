## Segmento 1: ¿Qué es una tabla de datos?

Una tabla organiza información en filas y columnas.

Por ejemplo, una tabla de alumnos puede tener columnas como:

- Nombre.
- Edad.
- Calificación.
- Hobby.
- Ciudad.

Cada fila representa un registro.

## Segmento 3: Crear nuestro primer `DataFrame`

Un `DataFrame` es una tabla en `pandas`.

Podemos crearlo a partir de un diccionario.

Un diccionario usa una estructura de clave y valores.

````python
import pandas as pd

datos = {
    "Nombre": ["Ana", "Luis", "Sofía", "Carlos"],
    "Edad": [14, 15, 14, 16],
    "Calificación": [9, 8, 10, 7]
}

tabla = pd.DataFrame(datos)

tabla
````

### Explicación del código

- `"Nombre"` es una columna.
- `"Edad"` es otra columna.
- `"Calificación"` es otra columna.
- Cada lista contiene los datos de esa columna.
- `pd.DataFrame(datos)` convierte los datos en tabla.

## Segmento 4: Ver las primeras filas

Cuando una tabla es grande, podemos usar `.head()` para ver solo las primeras filas.

````python
tabla.head()
````

## Segmento 5: Ver información general

`.info()` muestra información sobre la tabla.

- Cuántas filas hay.

````python
tabla.info()
````

## Segmento 6: Seleccionar una columna

Podemos seleccionar una columna escribiendo su nombre entre corchetes.

````python
tabla["Nombre"]
````

````python
tabla["Calificación"]
````

## Segmento 7: Calcular promedio

Si una columna tiene números, podemos calcular su promedio con `.mean()`.

````python
promedio = tabla["Calificación"].mean()

print("Promedio del grupo:", promedio)
````

## Segmento 8: Valor más alto y más bajo

Podemos usar:

- `.max()` para el valor más alto.
- `.min()` para el valor más bajo.

````python
print("Calificación más alta:", tabla["Calificación"].max())
print("Calificación más baja:", tabla["Calificación"].min())
````

## Segmento 9: Ordenar información

Podemos ordenar una tabla con `.sort_values()`.

Por ejemplo, ordenar de menor a mayor calificación.

````python
tabla.sort_values("Calificación")
````

Para ordenar de mayor a menor usamos:

```python
ascending=False
```

````python
tabla.sort_values("Calificación", ascending=False)
````

## Segmento 10: Filtrar datos

Filtrar significa mostrar solo los datos que cumplen una condición.

Ejemplo: mostrar alumnos con calificación mayor o igual a 8.

<div align="center">
  
</div>

````python
aprobados = tabla[tabla["Calificación"] >= 8]

aprobados
````

## Segmento 11: Filtrar por texto

También podemos filtrar usando texto.

Ejemplo: alumnos que tienen 14 años.

````python
alumnos_14 = tabla[tabla["Edad"] == 14]

alumnos_14
````

## Segmento 12: Agregar una nueva columna

Podemos crear columnas nuevas.

Por ejemplo, una columna que diga si el alumno aprobó.

````python
tabla["Aprobó"] = tabla["Calificación"] >= 6

tabla
````

La nueva columna muestra `True` o `False`.

- `True` significa verdadero.
- `False` significa falso.

## Segmento 13: Crear una tabla más completa

Ahora crearemos una tabla con más datos para practicar.

````python
import pandas as pd

datos = {
    "Nombre": ["Ana", "Luis", "Sofía", "Carlos", "Mariana", "Diego"],
    "Edad": [14, 15, 14, 16, 15, 14],
    "Materia favorita": ["Mate", "Historia", "Ciencias", "Arte", "Español", "Tecnología"],
    "Calificación": [9, 8, 10, 7, 9, 8],
    "Hobby": ["Dibujar", "Fútbol", "Leer", "Videojuegos", "Música", "Programar"]
}

alumnos = pd.DataFrame(datos)

alumnos
````

## Segmento 14: Preguntas que podemos responder

Con la tabla anterior podemos responder preguntas como:

- ¿Cuál es el promedio?
- ¿Quién tiene la calificación más alta?
- ¿Cuántos alumnos tienen 14 años?
- ¿Qué alumnos tienen calificación mayor o igual a 9?

````python
print("Promedio:", alumnos["Calificación"].mean())
print("Mayor calificación:", alumnos["Calificación"].max())
print("Menor calificación:", alumnos["Calificación"].min())
````

````python
alumnos_14 = alumnos[alumnos["Edad"] == 14]

alumnos_14
````

````python
mejores = alumnos[alumnos["Calificación"] >= 9]

mejores
````

## Segmento 15: Contar datos

Podemos usar `.value_counts()` para contar cuántas veces aparece cada valor.

Ejemplo: cuántos alumnos hay por edad.

````python
alumnos["Edad"].value_counts()
````

## Segmento 16: Crear gráfica desde `pandas`

`pandas` puede conectarse con `matplotlib` para crear gráficas.

Aquí graficaremos calificaciones por alumno.

````python
import matplotlib.pyplot as plt

alumnos.plot(x="Nombre", y="Calificación", kind="bar")

plt.title("Calificaciones por alumno")
plt.xlabel("Alumno")
plt.ylabel("Calificación")
plt.show()
````

## Segmento 17: Gráfica de conteos

Podemos contar edades y graficarlas.

````python
conteo_edades = alumnos["Edad"].value_counts()

conteo_edades.plot(kind="bar")

plt.title("Cantidad de alumnos por edad")
plt.xlabel("Edad")
plt.ylabel("Cantidad")
plt.show()
````
