## Segmento 3: Crear un botón

Un botón puede ejecutar una acción cuando el usuario hace clic.

```python
import ipywidgets as widgets
from IPython.display import display

boton = widgets.Button(description="Haz clic")

def accion(b):
    print("¡Botón presionado!")

boton.on_click(accion)

display(boton)
```

### Explicación del código

- `widgets.Button()` crea el botón.
- `description` define el texto del botón.
- `accion()` es la función que se ejecuta al hacer clic.
- `on_click()` conecta el botón con la función.
- `display()` muestra el botón.

## Segmento 4: Campo de texto

Podemos crear una caja para que el usuario escriba texto.

```python
import ipywidgets as widgets
from IPython.display import display

nombre = widgets.Text(
    description="Nombre:",
    placeholder="Escribe tu nombre"
)

display(nombre)
```

Después de escribir en la caja, podemos leer el valor con `.value`.

```python
print("Texto escrito:", nombre.value)
```

## Segmento 5: Botón + caja de texto

Ahora combinaremos una caja de texto con un botón.

```python
import ipywidgets as widgets
from IPython.display import display

nombre = widgets.Text(description="Nombre:")
boton = widgets.Button(description="Saludar")

def saludar(b):
    print("Hola", nombre.value)

boton.on_click(saludar)

display(nombre)
display(boton)
```

## Segmento 6: Slider

## Segmento 8: Mini app de recomendación

Vamos a crear una mini aplicación que recomienda una actividad según el clima.

```python
import ipywidgets as widgets
from IPython.display import display

clima = widgets.Dropdown(
    options=["soleado", "lluvioso", "nublado", "frío"],
    description="Clima:"
)

boton = widgets.Button(description="Recomendar")

def recomendar(b):
    if clima.value == "soleado":
        print("Recomendación: salir a caminar")
    elif clima.value == "lluvioso":
        print("Recomendación: ver una película")
    elif clima.value == "nublado":
        print("Recomendación: leer o dibujar")
    elif clima.value == "frío":
        print("Recomendación: tomar chocolate caliente")

boton.on_click(recomendar)

display(clima)
display(boton)
```

## Segmento 9: Mini app de calificaciones

Crearemos una app sencilla para calcular un promedio usando sliders.

```python
import ipywidgets as widgets
from IPython.display import display

cal1 = widgets.IntSlider(value=8, min=0, max=10, description="Cal 1:")
cal2 = widgets.IntSlider(value=8, min=0, max=10, description="Cal 2:")
cal3 = widgets.IntSlider(value=8, min=0, max=10, description="Cal 3:")

boton = widgets.Button(description="Calcular promedio")

def calcular(b):
    promedio = (cal1.value + cal2.value + cal3.value) / 3
    print("Promedio:", promedio)

boton.on_click(calcular)

display(cal1)
display(cal2)
display(cal3)
display(boton)
```

## Segmento 10: Mini app con gráfica

Podemos usar widgets para cambiar datos y luego graficarlos.

```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt

votos_python = widgets.IntSlider(value=5, min=0, max=20, description="Python")
votos_javascript = widgets.IntSlider(value=3, min=0, max=20, description="JS")
votos_html = widgets.IntSlider(value=4, min=0, max=20, description="HTML")

boton = widgets.Button(description="Crear gráfica")

def crear_grafica(b):
    temas = ["Python", "JavaScript", "HTML"]
    votos = [votos_python.value, votos_javascript.value, votos_html.value]

    plt.bar(temas, votos)
    plt.title("Votos por tema")
    plt.xlabel("Tema")
    plt.ylabel("Votos")
    plt.show()

boton.on_click(crear_grafica)

display(votos_python)
display(votos_javascript)
display(votos_html)
display(boton)
```

## Segmento 13: Opción de proyecto 1 — Quiz interactivo

Un quiz hace preguntas, cuenta puntos y muestra el resultado.

```python
def revisar_respuesta(respuesta, correcta):
    if respuesta.lower() == correcta.lower():
        return 1
    else:
        return 0

puntos = 0

respuesta = input("¿Qué lenguaje aprendimos en el curso? ")
puntos += revisar_respuesta(respuesta, "Python")

respuesta = input("¿Qué función muestra texto en pantalla? ")
puntos += revisar_respuesta(respuesta, "print")

respuesta = input("¿Qué biblioteca usamos para gráficas? ")
puntos += revisar_respuesta(respuesta, "matplotlib")

print("Puntaje final:", puntos, "de 3")
```

## Segmento 14: Opción de proyecto 2 — Simulador de calificaciones

Este proyecto captura calificaciones, calcula promedio y muestra una gráfica.

```python
import matplotlib.pyplot as plt

materias = []
calificaciones = []

for i in range(4):
    materia = input("Nombre de la materia: ")
    calificacion = float(input("Calificación: "))

    materias.append(materia)
    calificaciones.append(calificacion)

promedio = sum(calificaciones) / len(calificaciones)

print("Promedio:", promedio)

plt.bar(materias, calificaciones)
plt.title("Calificaciones")
plt.xlabel("Materia")
plt.ylabel("Calificación")
plt.show()
```

## Segmento 15: Opción de proyecto 3 — Encuesta del grupo

Este proyecto guarda respuestas y crea una gráfica.

```python
import matplotlib.pyplot as plt

opciones = ["Videojuegos", "Deporte", "Música", "Lectura"]
votos = [0, 0, 0, 0]

for i in range(5):
    print("Opciones:")
    print("1. Videojuegos")
    print("2. Deporte")
    print("3. Música")
    print("4. Lectura")

    voto = int(input("Elige una opción del 1 al 4: "))

    if voto >= 1 and voto <= 4:
        votos[voto - 1] += 1
    else:
        print("Voto no válido")

print("Resultados:", votos)

plt.bar(opciones, votos)
plt.title("Encuesta del grupo")
plt.xlabel("Opción")
plt.ylabel("Votos")
plt.show()
```

## Segmento 16: Opción de proyecto 4 — Generador de personajes

Este proyecto usa listas, funciones y `random`.

```python
import random

def elegir_nombre():
    nombres = ["Luna", "Max", "Nova", "Kai", "Leo"]
    return random.choice(nombres)

def elegir_poder():
    poderes = ["volar", "ser invisible", "controlar el fuego", "teletransportarse"]
    return random.choice(poderes)

def elegir_lugar():
    lugares = ["un castillo", "un planeta lejano", "una ciudad futurista", "un bosque mágico"]
    return random.choice(lugares)

def crear_personaje():
    print("----- PERSONAJE -----")
    print("Nombre:", elegir_nombre())
    print("Poder:", elegir_poder())
    print("Lugar:", elegir_lugar())

crear_personaje()
```

## Segmento 17: Opción de proyecto 5 — Mini app con widgets

Este proyecto usa controles interactivos.

```python
import ipywidgets as widgets
from IPython.display import display

nombre = widgets.Text(description="Nombre:")
edad = widgets.IntSlider(value=14, min=10, max=20, description="Edad:")
hobby = widgets.Dropdown(
    options=["Videojuegos", "Deporte", "Música", "Lectura", "Dibujo"],
    description="Hobby:"
)

boton = widgets.Button(description="Crear perfil")

def crear_perfil(b):
    print("----- PERFIL -----")
    print("Nombre:", nombre.value)
    print("Edad:", edad.value)
    print("Hobby:", hobby.value)

boton.on_click(crear_perfil)

display(nombre)
display(edad)
display(hobby)
display(boton)
```
