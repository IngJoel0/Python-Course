# Clase 4: Listas y manejo de datos

En esta clase aprenderemos a usar listas.

Una lista sirve para guardar varios datos dentro de una sola variable.

<div align="center">
  
</div>

## Segmento 1: ¿Qué es una lista?

Una lista es un grupo de datos.

En lugar de crear muchas variables separadas, podemos guardar todo junto.

```python
frutas = ["manzana", "plátano", "uva"]

print(frutas)
```

## Segmento 2: Listas de diferentes tipos

Podemos tener listas de textos, números o datos mezclados.

```python
nombres = ["Ana", "Luis", "Carlos"]
edades = [14, 15, 16]
datos = ["Sofía", 14, "Tijuana", True]

print(nombres)
print(edades)
print(datos)
```

## Segmento 3: Acceder a elementos

Cada elemento tiene una posición llamada índice.

Python empieza a contar desde 0.

```python
frutas = ["manzana", "plátano", "uva"]

print(frutas[0])
print(frutas[1])
print(frutas[2])
```

También podemos usar `-1` para obtener el último elemento.

```python
juegos = ["Minecraft", "Roblox", "Fortnite", "Mario Kart"]

print("Primer juego:", juegos[0])
print("Último juego:", juegos[-1])
```

## Segmento 4: Modificar una lista

Podemos cambiar un elemento indicando su posición.

```python
colores = ["rojo", "azul", "verde"]

print("Antes:", colores)

colores[1] = "morado"

print("Después:", colores)
```

## Segmento 5: Contar elementos con `len()`

`len()` dice cuántos elementos tiene una lista.

```python
peliculas = ["Spider-Man", "Shrek", "Toy Story", "Avatar"]

print("Cantidad:", len(peliculas))
```

## Segmento 6: Agregar elementos con `append()`

`append()` agrega un elemento al final de la lista.

<div align="center">
  
</div>

```python
amigos = ["Ana", "Luis"]

print("Antes:", amigos)

amigos.append("Carlos")

print("Después:", amigos)
```

## Segmento 7: Pedir datos y guardarlos

Podemos usar una lista vacía y llenarla con `input()`.

```python
canciones = []

for i in range(3):
    cancion = input("Escribe una canción favorita: ")
    canciones.append(cancion)

print("Tus canciones favoritas son:")
print(canciones)
```

## Segmento 8: Recorrer una lista

Recorrer una lista significa ver sus elementos uno por uno.

```python
frutas = ["manzana", "plátano", "uva"]

for fruta in frutas:
    print("Me gusta la", fruta)
```

Podemos mostrar una lista de forma más ordenada.

```python
tareas = ["Estudiar", "Practicar Python", "Leer"]

for tarea in tareas:
    print("-", tarea)
```

## Segmento 9: Recorrer con posición

A veces queremos mostrar el número de cada elemento.

```python
materias = ["Matemáticas", "Español", "Historia", "Ciencias"]

for i in range(len(materias)):
    print(i + 1, "-", materias[i])
```

## Segmento 10: Eliminar datos

Podemos eliminar por valor con `remove()` o por posición con `pop()`.

```python
animales = ["perro", "gato", "conejo", "pez"]

print("Antes:", animales)

animales.remove("conejo")

print("Después:", animales)
```

```python
personajes = ["Mario", "Luigi", "Peach", "Bowser"]

print("Antes:", personajes)

personajes.pop(1)

print("Después:", personajes)
```

## Segmento 11: Buscar dentro de una lista

Usamos `in` para saber si un elemento está en la lista.

```python
invitados = ["ana", "luis", "carlos", "sofia"]

nombre = input("Escribe tu nombre: ").lower()

if nombre in invitados:
    print("Puedes entrar")
else:
    print("No estás en la lista")
```

## Segmento 12: Ordenar listas

`sort()` ordena una lista.

```python
nombres = ["Sofía", "Ana", "Luis", "Carlos"]

print("Antes:", nombres)

nombres.sort()

print("Después:", nombres)
```

## Segmento 13: Listas con números

Con listas numéricas podemos calcular totales, promedios, máximos y mínimos.

```python
calificaciones = [8, 9, 10, 7, 9]

print("Total:", sum(calificaciones))
print("Cantidad:", len(calificaciones))
print("Promedio:", sum(calificaciones) / len(calificaciones))
print("Mayor:", max(calificaciones))
print("Menor:", min(calificaciones))
```

## Segmento 14: Registro de calificaciones

Ahora pediremos calificaciones al usuario.

```python
calificaciones = []

for i in range(5):
    calificacion = float(input("Escribe una calificación: "))
    calificaciones.append(calificacion)

promedio = sum(calificaciones) / len(calificaciones)

print("Calificaciones:", calificaciones)
print("Promedio:", promedio)
print("Más alta:", max(calificaciones))
print("Más baja:", min(calificaciones))
```

## Segmento 15: Listas aplicadas a juegos

Las listas también sirven para crear juegos o historias aleatorias.

```python
import random

personajes = ["robot", "mago", "explorador", "detective"]
lugares = ["castillo", "bosque", "planeta", "ciudad futurista"]

personaje = random.choice(personajes)
lugar = random.choice(lugares)

print("Tu historia será sobre un", personaje, "en un", lugar)
```
